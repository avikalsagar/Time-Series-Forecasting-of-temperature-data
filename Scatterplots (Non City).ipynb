```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
a = pd.read_csv('Point a.csv')
b = pd.read_csv('Point b.csv')
c = pd.read_csv('Point c.csv')
d = pd.read_csv('Point d.csv')
e = pd.read_csv('Point e.csv')
```

```python
X1 = a['Year']
y1 = a['Temperature (C)']
X2 = b['Year']
y2 = b['Temperature (C)']
X3 = c['Year']
y3 = c['Temperature (C)']
X4 = d['Year']
y4 = d['Temperature (C)']
X5 = e['Year']
y5 = e['Temperature (C)']
```

```python
sns.set_style("whitegrid")
fig, ax = plt.subplots(figsize=(15, 8))
plt.scatter(X1, y1)
plt.plot(X1,y1, label ='Point a',linestyle = '--')
plt.scatter(X2, y2, color='Red')
plt.plot(X2,y2, color='Red',label ='Point b',linestyle = '--')
plt.scatter(X3, y3, color='Hotpink')
plt.plot(X3,y3, color='Hotpink', label ='Point c',linestyle = '--')
plt.scatter(X4, y4, color='Orange')
plt.plot(X4,y4, color='Orange', label ='Point d',linestyle = '--')
plt.scatter(X5, y5, color='Green')
plt.plot(X5,y5, color='Green', label ='Point e',linestyle = '--')
plt.xticks(np.arange(min(X1), max(X1)+1, 1.0))
ax.get_xticklabels()[13].set_color("red")
ax.get_xticklabels()[14].set_color("red")
ax.get_xticklabels()[15].set_color("red")
ax.get_xticklabels()[16].set_color("red")
ax.get_xticklabels()[17].set_color("red")
ax.get_xticklabels()[18].set_color("red")
ax.get_xticklabels()[19].set_color("red")
ax.get_xticklabels()[20].set_color("red")
ax.get_xticklabels()[4].set_color("blue")
ax.get_xticklabels()[6].set_color("blue")
ax.get_xticklabels()[8].set_color("blue")
ax.get_xticklabels()[9].set_color("blue")
ax.get_xticklabels()[10].set_color("blue")
ax.get_xticklabels()[1].set_color("purple")
plt.plot([2014],[21.86],marker='D', markersize=15,color="black", alpha=0.3)
plt.plot([2019],[24.63],marker='D', markersize=15,color="black", alpha=0.3)
plt.xlabel('Year', size=22)
plt.ylabel('River Temperature (Â°C)', size=22)
plt.legend()
plt.savefig("Non River Temperature (City)",dpi=1000,bbox_inches = 'tight')
plt.show()
```
