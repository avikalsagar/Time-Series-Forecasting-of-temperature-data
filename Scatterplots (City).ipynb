```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patches as mpatches
```

```python
A = pd.read_csv('Point A.csv')
B = pd.read_csv('Point B.csv')
C = pd.read_csv('Point C.csv')
D = pd.read_csv('Point D.csv')
E = pd.read_csv('Point E.csv')
F = pd.read_csv('Point F.csv')
```

```python
X1 = A['Year']
y1 = A['Temperature (C)']
X2 = B['Year']
y2 = B['Temperature (C)']
X3 = C['Year']
y3 = C['Temperature (C)']
X4 = D['Year']
y4 = D['Temperature (C)']
X5 = E['Year']
y5 = E['Temperature (C)']
X6 = F['Year']
y6 = F['Temperature (C)']
```
